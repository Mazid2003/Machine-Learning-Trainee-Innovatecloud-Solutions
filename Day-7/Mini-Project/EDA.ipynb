Step 1: Import Required Libraries and Load Dataset



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a theme
sns.set_theme(style="whitegrid")

# Load dataset
titanic = sns.load_dataset("titanic")  # or use "titanic" if you want that instead

# View the data
print(titanic.head())
```

Step 2: Distribution of Numerical Features



```python
# Histogram for Age
sns.histplot(data=titanic, x='age', kde=True, bins=30)
plt.title("Distribution of Passenger Age")
plt.show()

# Histogram for Fare
sns.histplot(data=titanic, x='fare', kde=True, bins=30)
plt.title("Distribution of Fare Paid")
plt.show()
```

Step 3: Visualize Relationships Between Numeric Variables

```python
# Scatter plot: Age vs Fare
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived')
plt.title("Fare vs Age (Colored by Survival)")
plt.show()

# Correlation Heatmap
corr = titanic.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Titanic)")
plt.show()
```

Step 4: Analyze Categorical Breakdowns

```python
# Countplot of Survived
sns.countplot(data=titanic, x='survived')
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.show()

# Barplot of Survival Rate by Sex
sns.barplot(data=titanic, x='sex', y='survived')
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

Enhanced Visualization with Themes and Titles

```python
sns.set_theme(style="darkgrid", palette="Set2")

# Boxplot of Age by Pclass and Gender
sns.boxplot(data=titanic, x='pclass', y='age', hue='sex')
plt.title("Age Distribution Across Passenger Class and Gender")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.legend(title="Gender")
plt.show()
```
