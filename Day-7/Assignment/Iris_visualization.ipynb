Step 1. Importing necessary packages

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set a custom Seaborn theme
sns.set_theme(style="whitegrid", palette="muted")

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Preview the dataset
print(iris.head())
```

Step 2: Distribution Plots (2)


```python
# Distribution of sepal length
plt.figure(figsize=(8, 5))
sns.histplot(data=iris, x="sepal_length", kde=True, bins=20, color="blue")
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
plt.savefig("sepal_length_distribution.png")
plt.show()
```

```python
# Distribution of petal width
plt.figure(figsize=(8, 5))
sns.histplot(data=iris, x="petal_width", kde=True, bins=20, color="lightcoral")
plt.title("Distribution of Petal Width")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Count")
plt.savefig("petal_width_distribution.png")
plt.show()
```

Step 3: Relationship Plots (2)


```python
# sepal Length vs sepal Width
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", style="species", palette="deep")
plt.title("Sepal Length vs Sepal Width by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend(title="Species")
plt.savefig("sepal_length_vs_width.png")
plt.show()
```

```python
# petal length vs petal width
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species", style="species", palette="dark")
plt.title("Petal Length vs Petal Width by Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(title="Species")
plt.savefig("petal_length_vs_width.png")
plt.show()
```

Step 4: Categorical Plots (2)

```python
#Boxplot of petal length by species
plt.figure(figsize=(8, 5))
sns.boxplot(data=iris, x="species", y="petal_length", palette="pastel")
plt.title("Petal Length Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.savefig("boxplot_petal_length.png")
plt.show()
```

```python
#violen plot of sepal width by species
plt.figure(figsize=(8, 5))
sns.violinplot(data=iris, x="species", y="sepal_width", palette="Set2")
plt.title("Sepal Width Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.savefig("violinplot_sepal_width.png")
plt.show()
```
