Step 1: Load the CSV

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('weather_forecast_data.csv')

# Preview the data
print(df.head())
```

Line plot of temperature over days

```python
plt.figure(figsize=(8, 5))
plt.plot(df.index, df['Temperature'], marker='o', color='teal')
plt.title("Temperature Over Days")
plt.xlabel("Day Index")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()
```

Bar chart for average rainfall per month

```python
rain_count = df['Rain'].value_counts()

rain_count.plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Rain vs No Rain Days")
plt.xlabel("Weather")
plt.ylabel("Number of Days")
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

Pie chart for weather types (Sunny, Rainy, Cloudy)

```python
def classify_weather(row):
    if row['Rain'] == 'rain':
        return 'Rainy'
    elif row['Cloud_Cover'] > 50:
        return 'Cloudy'
    else:
        return 'Sunny'

df['Weather_Type'] = df.apply(classify_weather, axis=1)

weather_counts = df['Weather_Type'].value_counts()

# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(weather_counts, labels=weather_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Weather Type Distribution")
plt.axis('equal')
plt.show()
```

Scatter Plot

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Temperature'], df['Humidity'], color='purple')
plt.title('Temperature vs Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.grid(True)
plt.show()
```
