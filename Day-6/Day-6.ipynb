Basic Plotting
Example: Line chart

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

Scatter plot


```python
plt.scatter(x, y, color='blue', marker='.')
# for Bar chart
#plt.bar(categories, values, color='orange')
```

Pie Chart

```python
labels = ['Python', 'Java', 'C++']
sizes = [40, 30, 30]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Programming Language Popularity")
plt.axis('equal')
plt.show()
```

 Plot Customization Techniques:

1. color, linestyle, linewidth, marker

2. Adding legend(), title(), xlabel(), ylabel()

3. grid(True), xlim(), ylim()

Example

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]
plt.plot(x, y, color='red', label='Revenue', marker='s')
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Amount ($)")
plt.legend()
plt.grid(True)
plt.show()
```
